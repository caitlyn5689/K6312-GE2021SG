```python
import csv;
import numpy as np;
import matplotlib.pyplot as plt;
from datetime import datetime;
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
import matplotlib
from imblearn.over_sampling import SMOTE
from sklearn.decomposition import PCA

dicVariety = {}
dicYear = {}
dicTaster = {}
dicCountry = {}
dicProvince = {}
dicRegion = {}
```

```python
def loadTrainData(filePath):
    with open(filePath, encoding='utf-8') as csvfile:
        reader = csv.reader(csvfile)
        rows = [row for row in reader]
    # print(len(rows))  # 输出所有数据
    del (rows[0])

    varietyClass = 1
    yearClass = 1
    tasterClass = 1
    countryClass = 1
    provinceClass = 1
    regionClass = 1
    for row in rows:
        if row[1] not in dicVariety.keys():
            dicVariety[row[1]] = varietyClass
            row[1] = varietyClass
            varietyClass += 1
        else:
            row[1] = dicVariety[row[1]]

        if row[3] not in dicYear.keys():
            dicYear[row[3]] = yearClass
            row[3] = yearClass
            yearClass += 1
        else:
            row[3] = dicYear[row[3]]

        if row[4] not in dicTaster.keys():
            dicTaster[row[4]] = tasterClass
            row[4] = tasterClass
            tasterClass += 1
        else:
            row[4] = dicTaster[row[4]]

        if row[5] not in dicCountry.keys():
            dicCountry[row[5]] = countryClass
            row[5] = countryClass
            countryClass += 1
        else:
            row[5] = dicCountry[row[5]]

        if row[6] not in dicProvince.keys():
            dicProvince[row[6]] = provinceClass
            row[6] = provinceClass
            provinceClass += 1
        else:
            row[6] = dicProvince[row[6]]

        if row[7] not in dicRegion.keys():
            dicRegion[row[7]] = regionClass
            row[7] = regionClass
            regionClass += 1
        else:
            row[7] = dicRegion[row[7]]

        #转成数字格式
        for i in range(len(row)):
            row[i] = float(row[i])



    X = np.array(rows)#(8948, 16)
    Y = X[:,-1]#(8948,)
    X = X[:,1:-1]
    # print(X.shape)#(8948, 14)

    return X,Y
```

```python
trainX,trainY = loadTrainData("train_cleaned.csv")
testX, testY = loadTrainData("test_cleaned.csv")
```
